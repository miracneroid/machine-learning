## Importing Required Libraries

```python
# libraries required for data preprocessing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split

# libraries for standardizing the data
from sklearn.preprocessing import StandardScaler

# libraries for building the Neural Network
import tensorflow as tf 
tf.random.set_seed(3) #for getting same accuracy score as many times we train the model
from tensorflow import keras
```

## Data Collection and Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# loading the data to pandas data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data_frame
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
data_frame
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# getting some information about the data
data_frame.info()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

0 = Malignant tumor  ;  1 = Benign tumor

```python
# grouping the data based on the labels
data_frame.groupby('label').mean()
```

```python
# Separating the Features and Target
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

## Splitting the Data into Train (80% X_train) and Test (20% X_test)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

## Standardizing the Data

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

## Building and Training - the 'Neural Network' model
download.png

```python
# setting up the layers of Neural Network

model = keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')])

# input layer = keras.layers.Flatten()
# hidden layer = keras.layers.Dense(20, activation='relu')
# O\output Layer = keras.layers.Dense(2, activation='sigmoid')
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

### Saving the Model 

```python
import pickle
filename = 'Breast Cancer Classification Model_simple Neural Network.train'
pickle.dump(model, open(filename,'wb'))
#loadMODEL = pickle.load(open('Heart Disease Predictor Model.train','rb')) #loading the saved model
```

## Visualising the Accuraccy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

## Checking Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
